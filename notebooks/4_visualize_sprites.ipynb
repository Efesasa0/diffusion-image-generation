# DDPM

```python
file_path = '../outs/4/sprites_ddpm_file.pt'
```

```python
import torch
outputs = torch.load(file_path, map_location='cpu')
```

```python
import matplotlib.pyplot as plt

def show_images(imgs, nrow=2):
    _, axs = plt.subplots(nrow, imgs.shape[0] // nrow, figsize=(4,2 ))
    axs = axs.flatten()
    for img, ax in zip(imgs, axs):
        img = (img.permute(1, 2, 0).clip(-1, 1).detach().cpu().numpy() + 1) / 2
        ax.set_xticks([])
        ax.set_yticks([])
        ax.imshow(img)
    plt.show()

show_images(outputs)
```
